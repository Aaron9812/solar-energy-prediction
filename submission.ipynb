## 1.2 Dataset Overview <a id='overview'></a>

## 1.3 Feature reduction <a id='feature-reduction'></a>
Given the large number of features in the dataset, some of which may be difficult to interpret without domain knowledge in meteorology, it can be challenging to identify meaningful relationships among the variables. Therefore, in order to facilitate model interpretability and potentially improve prediction performance, we propose reducing the number of features. Specifically, we will focus on selecting features that have been previously identified as having a significant impact on model performance, as discussed in relevant lecture. This process will allow us to mitigate the "curse of dimensionality" and improve the efficiency of our modeling efforts.


We propose excluding certain columns from the dataset for the following reasons:

- <b>"WindComponentUat0, WindComponentVat0, WindComponentUat100, and WindComponentVat100"</b> will be excluded as their meaning and impact on power output prediction are unclear. Instead, we will rely on the more easily interpretable LowerWindSpeed and UpperWindSpeed columns for information about wind speed.

- <b>"clearsky_diffuse, clearsky_direct, clearsky_global, clearsky_diffuse_agg, clearsky_direct_agg, and clearsky_global_agg"</b> will be excluded due to their potential difficulty in interpretation and dependence on cloud cover. Cloud cover, which is measured separately in the TotalCloudCoverAt0 column, can be a more informative metric in predicting solar power output.

- <b>"RelativeHumidityAt1000 and RelativeHumidityAt950"</b> columns, while containing potentially valuable information, may be difficult for the model to interpret due to the use of three different pressure measurements at varying altitudes. Therefore, we will focus solely on the RelativeHumidityAt0 column, which is measured at ground level.

- <b>"PotentialVorticityAt950"</b> will be excluded since the PotentialVorticityAt1000 measurement should be sufficient for modeling purposes.



```python
# drop indComponentUat0, WindComponentVat0, WindComponentUat100, and WindComponentVat100
df = df.drop(['WindComponentUat0', 'WindComponentVat0', 'WindComponentUat100', 'WindComponentVat100'], axis=1)
# drop clearsky_diffuse, clearsky_direct, clearsky_global, clearsky_diffuse_agg, clearsky_direct_agg, and clearsky_global_agg
df = df.drop(['clearsky_diffuse', 'clearsky_direct', 'clearsky_global', 'clearsky_diffuse_agg', 'clearsky_direct_agg', 'clearsky_global_agg'], axis=1)
# drop RelativeHumidityAt1000 and RelativeHumidityAt950
df = df.drop(['RelativeHumidityAt1000', 'RelativeHumidityAt950'], axis=1)
# drop PotentialVorticityAt950
df = df.drop(['PotentialVorticityAt950'], axis=1)
```

# 2. Linear regression baseline model <a id='baseline'></a>

```python
def eval(model, X_test: pd.DataFrame, y_test: pd.DataFrame)-> dict:
    # Predict the target variable for the test data
    y_pred = model.predict(X_test)

    # Clip the predicted values to a minimum of 0
    y_pred_clipped = np.clip(y_pred, a_min=0, a_max=None)
    
    # Calculate evaluation metrics
    mse = mean_squared_error(y_test, y_pred_clipped)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    return {
        "mse": mse,
        "rmse": rmse,
        "r2": r2
    }
```

```python
# Get data
X_train, y_train, X_test, y_test = utlis.split_data(df_reduced=df_reduced, split_rate=0.8)

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

results = eval(model, X_test, y_test)

print(f"Mean Squared Error: {results.get('mse'):.4f}")
print(f"Root Mean Squared Error: {results.get('rmse'):.4f}")
print(f"R-squared: {results.get('r2'):.4f}")
```

- **Mean Squared Error (MSE) - 0.0037:** The MSE measures the average squared difference between the actual and predicted target values. A lower value indicates better performance, with 0 being a perfect prediction. In this case, the MSE of 0.0038 suggests that the model has relatively small errors in its predictions, indicating a reasonably good fit.

- **Root Mean Squared Error (RMSE) - 0.0611:** The RMSE represents the square root of the MSE and is expressed in the same units as the target variable. It provides an estimate of the average deviation of the model's predictions from the actual values. An RMSE of 0.0615 implies that, on average, the model's predictions deviate from the actual target values by approximately 6.15% (assuming the target variable is expressed as a ratio or percentage). This value indicates a moderately accurate prediction performance.

- **R-squared - 0.8628:** R-squared is a measure of how well the model's predictions fit the actual data, ranging from 0 to 1. A higher value is better, with 1 indicating that the model perfectly predicts the target variable. An R-squared of 0.8628 means that approximately 86.28% of the variance in the target variable can be explained by the model. This relatively high R-squared value suggests that the model has a good fit to the data.

**Note:** The predictions were clipped below 0 as values below 0 dont make a lot of sense. 



```python
# Lets compare the results of the Linear Regression model to the acctual values
# Get the predictions for the test data
y_pred = model.predict(X_test)

# Clip the predicted values to a minimum of 0
y_pred_clipped = np.clip(y_pred, a_min=0, a_max=None)

# Create a dataframe with the actual and predicted values
df_pred = pd.DataFrame({'Actual': y_test.values.flatten(), 'Predicted': y_pred_clipped.flatten(), "Actual - Predicted": y_test.values.flatten() - y_pred_clipped.flatten()})

# Plot the actual and predicted values
fig = px.line(df_pred, x=df_pred.index, y=['Actual', 'Predicted', 'Actual - Predicted'])
fig.update_layout(
    xaxis_title="Time",
    yaxis_title="Normalized Power",
    legend_title="Actual vs Predicted",
    title="Normalized Power of a Solar Farm over Time",
    font=dict(
        family="Montserrat, monospace",
        size=12,
        color="#7f7f7f"
    )
)
fig.show()
```

The initial model performance shows promising results as a starting point. Although there is no distinct weakness such as a general tendency to overpredict or underpredict the actual results.

```python
# Lets look at the weights of the model and see which features are most important
# Get the weights of the model
weights = model.coef_

# Get the feature names
feature_names = X_train.columns

# Create a dataframe with the weights and feature names
df_weights = pd.DataFrame({'Feature': feature_names, 'Weight': weights})

# Sort the dataframe by the weights
df_weights = df_weights.sort_values(by='Weight', ascending=False)

# Plot the weights
fig = px.bar(df_weights, x='Weight', y='Feature', orientation='h')
fig.update_layout(
    xaxis_title="Weight",
    yaxis_title="Feature",
    legend_title="",
    title="Feature Weights",
    font=dict(
        family="Montserrat, monospace",
        size=12,
        color="#7f7f7f"
    )
)
fig.show()
```

Additionally, it is intriguing to observe the significance attributed to the "dewpoint temperature" feature by the model, despite having a nearly identical correlation coefficient to that of the "hour of the day" feature. 

An analysis of its generalization performance is needed to get a better idea of the model's performance. For this purpose, it should evaluate every Dataset.

```python
def eval_multiple_files(model):
    results = []
    for i in range(21):
        # open csv file
        filename = f"GermanSolarFarm/data/pv_{str((i+1)).zfill(2)}.csv"
        df = pd.read_csv(filename, delimiter=';')
        df_reduced = utlis.reduce_dataframe(df=df)
        X_train, y_train, X_test, y_test = utlis.split_data(df_reduced=df_reduced, split_rate=0.8)

        results.append({
            "file": f"pv_{i+1}",
            "results": eval(model, X_test, y_test)
            })
    return results
```

```python
results_linear = eval_multiple_files(model)
# calulate averages
avg_mse, avg_rmse, avg_r2 = utlis.calculate_average_performance(results_linear)

print(f"Average mean Squared Error: {avg_mse:.4f}")
print(f"Average root Mean Squared Error: {avg_rmse:.4f}")
print(f"Average R-squared: {avg_r2:.4f}")
```

```python
# Extract the filenames and the results into separate lists
filenames = [item["file"] for item in results_linear]
mre_values = [item["results"]["mse"] for item in results_linear]
r2_values = [item["results"]["r2"] for item in results_linear]
rmse_values = [item["results"]["rmse"] for item in results_linear]

# Create a grouped bar chart using Plotly
fig = go.Figure(data=[
    go.Bar(name='MSE', x=filenames, y=mre_values),
    go.Bar(name='RMSE', x=filenames, y=rmse_values)
])

# Customize the layout
fig.update_layout(
    barmode='group',
    title='Results for Different Files',
    xaxis_title='Filename',
    yaxis_title='Value',
    legend_title='Metrics',
)

# Show the plot
fig.show()
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import SVR, LinearSVR
from sklearn.metrics import mean_squared_error, r2_score
import utlis
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
import xgboost as xgb

# Get data
X_train, y_train, X_test, y_test = utlis.split_data(df_reduced=df_reduced, split_rate=0.8)

nb_features = len(X_train.columns.tolist())
nb_targets = 1
layer_size = nb_features + nb_targets + 2

# Define the base models to use in the ensemble
base_models = [
    ('lr', LinearRegression()),
    ('ridge', Ridge(random_state=seed)),
    ('lasso', Lasso(random_state=seed)),
    ('dt_friedman_mse', DecisionTreeRegressor(criterion='friedman_mse', random_state=seed)),
    ('dt_absolute_error', DecisionTreeRegressor(criterion='absolute_error', random_state=seed)),
    ('dt_poisson', DecisionTreeRegressor(criterion='poisson', random_state=seed)),
    ('RFRS',RandomForestRegressor(criterion='squared_error', n_estimators=100, random_state = seed)), 
    ('RFRA', RandomForestRegressor(criterion='absolute_error', n_estimators=100, random_state = seed)), 
    ('RFRP', RandomForestRegressor(criterion='poisson', n_estimators=100, random_state = seed)), 
    ('mlp', MLPRegressor(random_state=seed)),
    ('ada', AdaBoostRegressor(random_state=seed)),
    ('knr', KNeighborsRegressor()),
    ('gpr', GaussianProcessRegressor(random_state=seed)),
    ('MLPR1', MLPRegressor(hidden_layer_sizes = (layer_size, ), max_iter=2000, early_stopping=True, random_state = seed)), 
    ('MLPR2', MLPRegressor(hidden_layer_sizes = (layer_size, layer_size,), max_iter=2000, early_stopping=True, random_state = seed)), 
    ('elastic_net', ElasticNet(alpha=0.01, l1_ratio=0.15, random_state=seed)),
    ('XGBR', xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1)),
    ('LSVR', LinearSVR(epsilon=11, C=37, fit_intercept=True, loss='squared_epsilon_insensitive', max_iter=30_000, random_state=seed)),
    ('SVR', SVR(kernel='poly', C=1e3, gamma=0.1, epsilon=0.1))
]

# Define the meta-model to use in the ensemble
meta_model = LinearRegression()

# Initialize the Stacking Regressor
stacking_model = StackingRegressor(estimators=base_models, final_estimator=meta_model, n_jobs=-1)

# Fit the Stacking Regressor to the training data
stacking_model.fit(X_train, y_train)

# Make predictions on the test set using the Stacking Regressor
y_pred = stacking_model.predict(X_test)

# Evaluate the performance of the Stacking Regressor using mean squared error, root mean squared error, and R-squared
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R-squared: {r2:.4f}")
```
